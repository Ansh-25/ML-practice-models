```python
import math

def sqaure_check(n):
  root = math.sqrt(n)
  if int(root + 0.5) ** 2 == n:
    return True
  else:
    return False

numbers=[3]
lis=[]
exception=[]

for i in range(5,7000,2):
  flag = 0
  for j in numbers:
    if i%j == 0:
      flag = 1
  if flag == 0 :
    numbers.append(i)
  else:
    for k in numbers:
      check = int((i-k)/2)
      if (i-k)%2==0:
        if sqaure_check(check):
          lis.append((i,k,int(math.sqrt((i-k)/2))))
          flag = 2
        else:
          continue
    if flag != 2 :
      exception.append(i)
      break 
    
print(lis)
print(exception)
```
