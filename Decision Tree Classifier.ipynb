# Decision Tree Classifier 

```python
%matplotlib inline

import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn import tree
import pandas as pd
import numpy as np
```

```python
# Read csv file and saperate columns
music_data = pd.read_csv('./Datasets/music.csv')
X = music_data.drop(columns =['genre'])
y = music_data['genre']

music_data.head()
```

```python
# Split data into train and test
X_train, X_test , y_train , y_test = train_test_split(X,y,test_size=0.2)
```

```python
# instantiate and fit the model
model = DecisionTreeClassifier()
model.fit(X_train, y_train)
```

```python
# test prediction
predictions = model.predict(X_test)
print(predictions)
```

```python
# accuracy score
score = accuracy_score(y_test, predictions)
score
```

[plot_tree documentation](https://scikit-learn.org/stable/modules/generated/sklearn.tree.plot_tree.html)

```python
# visulaizing decision tree
fig = plt.figure(figsize=(25,10))
_ = tree.plot_tree(model, 
                   feature_names=['age','gender'],  
                   class_names=sorted(y.unique()),
                   label ='all',rounded = True,
                   filled=True)
```
